```python
import os
import numpy as np
import pandas as pd
dataframe = pd.read_csv('/home/yupaporn/codes/USAI/FVtrain_fold5_1.csv') #เปลี่ยนไฟล์
dataframe = dataframe.drop(['Unnamed: 0'], axis=1)
dataframe
```

```python
X = dataframe.iloc[:,0:2048]
y = dataframe.iloc[:,2051]
print(X.shape)
print(y.shape)
```

```python
y[0]
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import BaggingClassifier
```

# setting parameters

```python
forestOpt = RandomForestClassifier(random_state = 1, max_depth = 5, n_estimators = 400, min_samples_split =2) #เปลี่ยนตาม cell 8
             
modelOpt = forestOpt.fit(X, y)# 
```

```python
dataframe = pd.read_csv( '/home/yupaporn/codes/USAI/FVval_fold5_1.csv') #เปลี่ยนชื่อไฟล์ Validation
```

```python
dataframe
```

```python
X_test = dataframe.iloc[:,1:2049]
y_test = dataframe.iloc[:,2052]
print(X_test.shape)
print(y_test.shape)
```

```python
y_test
```

```python
y_pred = modelOpt.predict(X_test)
```

```python
#Marking the Confusion Matrix
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))#performance
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
cm = confusion_matrix(y_test, y_pred)

import seaborn as sns
sns.set()
fig, ax = plt.subplots(figsize=(8, 5))

ax = sns.heatmap(cm, annot=True, fmt="d", cmap="YlGnBu") #Blues,Oranges,Reds
ax.set_title('Confusion matrix',fontsize=20)
ax.set_ylabel('True label',fontsize=18)
ax.set_xlabel('Predicted label',fontsize=18)
```
