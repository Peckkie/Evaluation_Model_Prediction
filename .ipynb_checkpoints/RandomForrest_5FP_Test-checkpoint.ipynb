```python
import os
import numpy as np
import pandas as pd
#load model
import pickle
os.chdir('/media/tohn/SSD/ModelTrainByImages/R2_1/models/')
filename = "modelrf_EffNet5_b5_1FC_cross_15AB_5FP.pkl"
with open(filename, 'rb') as file:
    model = pickle.load(file)
```

```python
dataframe = pd.read_csv( '/home/yupaporn/codes/USAI/FeaturesVector_Testset_15AB.csv')
dataframe = dataframe.drop(['Unnamed: 0'], axis=1)
dataframe
```

```python
X_test = dataframe.iloc[:,0:2048]
y_test = dataframe.iloc[:,2050]
print(X_test.shape)
print(y_test.shape)
```

```python
X_test
```

```python
y_test
```

```python
y_pred = model.predict(X_test)
```

```python
y_prob =model.predict_proba(X_test)
```

```python
y_probmax =[]
for i in range(len(y_prob)):
    a = y_prob[i]
    a = a[np.argmax(a)]
    y_probmax.append(a)
```

```python
y_prob.tolist()
```

```python
df0 = pd.read_csv (r'/home/yupaporn/codes/USAI/Testdf_result.csv')
df0['FP_category'] = y_pred
df0['FP_Prob'] = y_probmax
df0['FP_ProbAll'] = y_prob.tolist()

os.chdir( '/home/yupaporn/codes/USAI/')
df0.to_csv('Testdf_result.csv')
```
